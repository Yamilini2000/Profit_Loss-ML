```python
#IMPORTING LIBRARIES
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#IMPORTING DATASET
dataset = pd.read_csv('D:/DATASETS/Random/ProfitLoss.csv')
dataset
```

```python
#INITIALISING INDEPENDENT VARIABLE & DEPENDENT VARIABLE
X = dataset.iloc[:, 2:4].values   
y = dataset.iloc[:, 5].values
print("X{} \n \n ... \n \n y{}".format(X, y))
```

```python
#NORMALIZING y
for i in range(0,500):
    if y[i] <= 0:
        y[i] = 0        # 0 for loss
    else:
        y[i] = 1        #1 for profit
y
```

```python
#SPLITTING DATA INTO TRAINING AND TESTING SET
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size = 0.2, random_state = 0)
```

```python
#FEATURING SCALING
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
#FITTING DECISION TREE TO TRAINING SET
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
classifier.fit(X_train, y_train)
```

```python
#PREDICTING TEST SET
y_pred = classifier.predict(X_test)
y_pred
```

```python
#CREATING CONFUSION MATRIX
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
#VISUALIZING TRAINING SET RESULTS
from matplotlib.colors import ListedColormap
X_set, y_set = X_train, y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Profit v/s Loss(Training set)')
plt.xlabel('Profit Markup')
plt.ylabel('Depth of Tree')
plt.legend()
plt.show()
```

```python
#VISUALIZING TEST SET RESULTS
from matplotlib.colors import ListedColormap
X_set, y_set = X_test, y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Profit v/s Loss(Test set)')
plt.xlabel('Profit Markup')
plt.ylabel('Depth of Tree')
plt.legend()
plt.show()
```

```python
#COMPUTING ACCURACY
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
accuracy
```
